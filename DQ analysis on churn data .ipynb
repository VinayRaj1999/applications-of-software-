### Summarize DQ Findings

```python
summary=vr1.describe()
print(summary)
```

```python
iqr_report = {}
```

```python
for column in vr1.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = vr1[column].quantile(0.25)
    Q3 = vr1[column].quantile(0.75)
    IQR = Q3 - Q1
    iqr_report[column] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': Q1 - 1.5 * IQR,
        'Upper Bound': Q3 + 1.5 * IQR}
```

```python
iqr_report_df = pd.DataFrame.from_dict(iqr_report, orient='index')
```

```python
print(iqr_report_df)
```
