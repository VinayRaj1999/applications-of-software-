```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the provided file path
file_path = "C://Users//kiran//Downloads//All_States.xlsx"
df = pd.read_excel(file_path)

# Filter data for 2014-15
df_2014_15 = df[df['Item'] == '2014-15']

# Remove union territories
union_territories = ['Delhi', 'Chandigarh', 'Andaman and Nicobar Islands', 
                     'Lakshadweep', 'Dadra and Nagar Haveli and Daman and Diu', 'Puducherry']
df_states = df_2014_15[~df_2014_15['State'].isin(union_territories)]

# Display the cleaned data
print(df_states.head())
```

```python
# Step 2: GDP per capita analysis
# Plot GDP per capita
plt.figure(figsize=(12, 6))
df_states = df_states.sort_values(by='Per Capita GSDP (Rs.)', ascending=False)
sns.barplot(data=df_states, x='State', y='Per Capita GSDP (Rs.)', palette='viridis')
plt.xticks(rotation=90)
plt.title('GDP Per Capita (2014-15)')
plt.ylabel('Per Capita GDP (Rs.)')
plt.xlabel('State')
plt.show()
```

```python
# Identify top and bottom 5 states
top_5_per_capita = df_states.nlargest(5, 'Per Capita GSDP (Rs.)')[['State', 'Per Capita GSDP (Rs.)']]
bottom_5_per_capita = df_states.nsmallest(5, 'Per Capita GSDP (Rs.)')[['State', 'Per Capita GSDP (Rs.)']]
print("Top 5 States by GDP Per Capita:\n", top_5_per_capita)
print("Bottom 5 States by GDP Per Capita:\n", bottom_5_per_capita)

# Ratio of highest to lowest GDP per capita
ratio = top_5_per_capita.iloc[0]['Per Capita GSDP (Rs.)'] / bottom_5_per_capita.iloc[0]['Per Capita GSDP (Rs.)']
print(f"Ratio of highest to lowest GDP per capita: {ratio:.2f}")
```

```python
# Step 3: Sectoral contribution analysis
# Compute sectoral contributions as percentages
df_states['Primary (%)'] = (df_states['Primary'] / df_states['TOTAL GSVA at basic prices']) * 100
df_states['Secondary (%)'] = (df_states['Secondary'] / df_states['TOTAL GSVA at basic prices']) * 100
df_states['Tertiary (%)'] = (df_states['Tertiary'] / df_states['TOTAL GSVA at basic prices']) * 100

# Plot sectoral contributions
sector_data = df_states[['State', 'Primary (%)', 'Secondary (%)', 'Tertiary (%)']].melt(
    id_vars='State', var_name='Sector', value_name='Contribution (%)'
)

plt.figure(figsize=(12, 6))
sns.barplot(data=sector_data, x='State', y='Contribution (%)', hue='Sector')
plt.xticks(rotation=90)
plt.title('Sectoral Contribution to GDP (2014-15)')
plt.ylabel('Contribution (%)')
plt.xlabel('State')
plt.legend(title='Sector')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Compute sectoral contributions as percentages
df_states['Primary (%)'] = (df_states['Primary'] / df_states['TOTAL GSVA at basic prices']) * 100
df_states['Secondary (%)'] = (df_states['Secondary'] / df_states['TOTAL GSVA at basic prices']) * 100
df_states['Tertiary (%)'] = (df_states['Tertiary'] / df_states['TOTAL GSVA at basic prices']) * 100

# Plot Primary sector contribution in descending order
df_primary_sorted = df_states.sort_values('Primary (%)', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_primary_sorted, x='State', y='Primary (%)', palette='Blues_d')
plt.xticks(rotation=90)
plt.title('Primary Sector Contribution to GDP (2014-15)')
plt.ylabel('Primary Sector Contribution (%)')
plt.xlabel('State')
plt.show()

# Plot Secondary sector contribution in descending order
df_secondary_sorted = df_states.sort_values('Secondary (%)', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_secondary_sorted, x='State', y='Secondary (%)', palette='Greens_d')
plt.xticks(rotation=90)
plt.title('Secondary Sector Contribution to GDP (2014-15)')
plt.ylabel('Secondary Sector Contribution (%)')
plt.xlabel('State')
plt.show()

# Plot Tertiary sector contribution in descending order
df_tertiary_sorted = df_states.sort_values('Tertiary (%)', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_tertiary_sorted, x='State', y='Tertiary (%)', palette='Purples_d')
plt.xticks(rotation=90)
plt.title('Tertiary Sector Contribution to GDP (2014-15)')
plt.ylabel('Tertiary Sector Contribution (%)')
plt.xlabel('State')
plt.show()
```

```python
# Step 4: Categorize states into quantile groups
# Categorize states based on GDP per capita
quantiles = df_states['Per Capita GSDP (Rs.)'].quantile([0.20, 0.5, 0.85, 1]).values
df_states['Category'] = pd.cut(
    df_states['Per Capita GSDP (Rs.)'],
    bins=[-np.inf, quantiles[0], quantiles[1], quantiles[2], np.inf],
    labels=['C4', 'C3', 'C2', 'C1']
)
print(df_states[['State', 'Per Capita GSDP (Rs.)', 'Category']])
```

```python
# Step 5: Top sub-sectors contributing to 80% of GSDP
# Calculate sub-sector contributions
sub_sectors = ['Agriculture, forestry and fishing', 'Crops', 'Livestock', 'Forestry and logging', 
               'Fishing and aquaculture', 'Mining and quarrying', 'Manufacturing', 
               'Electricity, gas, water supply & other utility services', 'Construction', 
               'Trade, repair, hotels and restaurants', 'Transport, storage, communication & services related to broadcasting', 
               'Financial services', 'Real estate, ownership of dwelling & professional services', 
               'Public administration', 'Other services']

# Analyze each category
for category in ['C1', 'C2', 'C3', 'C4']:
    print(f"\nCategory {category} Analysis:")
    category_data = df_states[df_states['Category'] == category]
    sub_sector_contributions = category_data[sub_sectors].sum().sort_values(ascending=False)
    total_contribution = sub_sector_contributions.sum()
    sub_sector_percentages = (sub_sector_contributions / total_contribution) * 100
    top_sub_sectors = sub_sector_percentages.cumsum()
    
    # Identify top sub-sectors contributing ~80%
    top_80 = sub_sector_percentages[top_sub_sectors <= 80]
    print("Top Sub-Sectors Contributing to 80% of GSDP:\n", top_80)
    
    # Plot top sub-sectors
    plt.figure(figsize=(8, 5))
    top_80.plot(kind='bar', color='skyblue')
    plt.title(f'Top Sub-Sectors for Category {category}')
    plt.ylabel('Contribution to GSDP (%)')
    plt.xlabel('Sub-Sectors')
    plt.show()
```

```python
# Step 6: Derive insights and recommendations
# Example recommendations framework
for category in ['C1', 'C2', 'C3', 'C4']:
    print(f"\nRecommendations for Category {category}:")
    if category == 'C1':
        print("1. Invest in high-value services and advanced manufacturing.")
        print("2. Enhance export-oriented industries.")
    elif category == 'C2':
        print("1. Strengthen infrastructure for secondary sectors like construction.")
        print("2. Support agriculture-based industries for higher value addition.")
    elif category == 'C3':
        print("1. Focus on improving tertiary services like trade and communication.")
        print("2. Expand access to financial services.")
    elif category == 'C4':
        print("1. Boost primary sectors like agriculture with modern techniques.")
        print("2. Develop basic manufacturing and small industries.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (adjust the file path if necessary)
file_path = "C://Users//kiran//Downloads//All_States.xlsx"

# Load the data from the sheet (assuming all data is in the first sheet)
df = pd.read_excel("C://Users//kiran//Downloads//All_States.xlsx")

# Check the column names to identify relevant data
print(df.columns)

# Assuming the columns for GDP per capita and dropout rates are like:
# 'Per Capita GSDP (Rs.)', 'Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout'

# Drop rows with missing data in 'Per Capita GSDP (Rs.)' and dropout rates columns
df_cleaned = df.dropna(subset=['Per Capita GSDP (Rs.)', 'Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout'])

# Calculate the correlation between Per Capita GSDP and dropout rates (for each education level)
correlation_matrix = df_cleaned[['Per Capita GSDP (Rs.)', 'Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout']].corr()

# Plotting the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix: Per Capita GSDP vs. Dropout Rates')
plt.show()

# Scatter plots to visualize the relationship
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Primary education dropout vs GDP per capita
sns.scatterplot(x='Per Capita GSDP (Rs.)', y='Primary Dropout', data=df_cleaned, ax=axes[0])
axes[0].set_title('Primary Education Dropout vs GDP per Capita')

# Upper primary education dropout vs GDP per capita
sns.scatterplot(x='Per Capita GSDP (Rs.)', y='Upper Primary Dropout', data=df_cleaned, ax=axes[1])
axes[1].set_title('Upper Primary Education Dropout vs GDP per Capita')

# Secondary education dropout vs GDP per capita
sns.scatterplot(x='Per Capita GSDP (Rs.)', y='Secondary Dropout', data=df_cleaned, ax=axes[2])
axes[2].set_title('Secondary Education Dropout vs GDP per Capita')

plt.tight_layout()
plt.show()
```
